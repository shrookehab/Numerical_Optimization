## LR Full Implementation

### Step1: Initialize parameters (theta_0 & theta_1) with random value or simply zero. Also choose the Learning rate. 

```python
theta = [0,0]
learning_rate = 0.001
m = len(x_points)
```

### Step2: Use (theta_0 & theta_1) to predict the output h(x)= theta_0 + theta_1 * x.
#### Note: you will need to iterate through all data points

```python
hThetaX = np.add(theta[0], np.multiply(theta[1], x_points))
```

### Step3: Calculate Cost function (theta_0,theta_1 ).


```python
JTheta = (1 / 2*len(x_points)) * np.sum(np.power(np.subtract(hThetaX, y_points), 2))
```

### Step4: Calculate the gradient.


```python
GDTheta0 = (1 / len(x_points)) * np.sum(np.multiply(np.subtract(hThetaX, y_points), 1))
GDTheta1 = (1 / len(x_points)) * np.sum(np.multiply(np.subtract(hThetaX, y_points), x_points))
```

### Step5: Update the parameters (simultaneously).


```python
theta[0] = theta[0] - learning_rate * GDTheta0
theta[1] = theta[1] - learning_rate * GDTheta1
```

### Step6: Repeat from 2 to 5 until converge to the minimum or achieve maximum iterations.

```python
for i in range(1000):
    hThetaX = np.add(theta[0], np.multiply(theta[1], x_points))
    #JTheta = (1 / 2 * m) * np.sum(np.power(np.subtract(hThetaX, y_points), 2))
    JTheta = (1 / (2 * m)) * np.sum(np.subtract(hThetaX, y_points) **2)
    GDTheta0 = (1 / m) * np.sum(np.multiply(np.subtract(hThetaX, y_points), 1))
    GDTheta1 = (1 / m) * np.sum(np.multiply(np.subtract(hThetaX, y_points), x_points))
    theta[0] = theta[0] - learning_rate * GDTheta0
    theta[1] = theta[1] - learning_rate * GDTheta1
    print('error using MSE = ', JTheta)
    print('New Theta = ', theta)
    print('')
```

#### Predict y values using the LR equation
##### h(x)= theta_0 + theta_1 * x

```python
hThetaX = np.add(theta[0], np.multiply(theta[1], x_points))
print(hThetaX)
```

#### Plot  LR equation output (fitted line) with the original data (scatter plot of X,y)

```python
plt.scatter(x_points, y_points)
z = np.polyfit(x_points, hThetaX, 1)
p = np.poly1d(z)
plt.plot(x_points,p(x_points))
plt.title('relation between x_points and y_points')
plt.ylabel('y_points')
plt.xlabel('x_points')
plt.show()
```

### Repeat your implementation but for more than one variable

```python
def GDBatchMulVar(x_points, y_points, learning_rate, maxNumOfIteration):
    m = len(x_points)
    theta = np.zeros((4,1))
    loss = []
    thetas0 = []
    thetas1 = []
    thetas2 = []
    thetas3 = []
    hyposesis = []
    iteration = 1
    numIter = []
    GDthetas = np.zeros((4,1))
    for i in range(maxNumOfIteration):
        print('==================iteration{}==================='.format(iteration))
        hThetaX = (x_points @ theta)
        hyposesis.append(hThetaX)
        numIter.append(i)
        #JTheta = (1 / 2 * m) * np.sum(np.power(np.subtract(hThetaX, y_points), 2))
        error = np.subtract(hThetaX, y_points)
        JTheta = (np.transpose(error)@ error) / (2 * m)
        loss.append(JTheta)
        GDTheta = (np.transpose(x_points)@ error) / m
        thetas0.append(theta[0])
        thetas1.append(theta[1])
        thetas2.append(theta[2])
        thetas3.append(theta[3])
        theta = np.subtract(theta, np.multiply(learning_rate, GDTheta))
        GDthetas = GDTheta
        print('error using MSE = ', JTheta)
        print('New Theta = ', theta)
        print('')
        iteration = iteration + 1
        #norm = L2Norm(gradTheta)
        #norm = np.absolute(np.dot(GDTheta0, GDTheta1))
        norm = np.linalg.norm(GDthetas)
        if i > 0:
            if np.absolute(loss[i] - loss[i-1]) < 0.001 or norm < 1:
                break
        optimum = (x_points @ theta)
#print('loss is: ', loss)
#print('thetas is: ', theta)
    return [thetas0, thetas1, thetas2, thetas3, loss, hyposesis, optimum, numIter]
```

#### Predict y values using the LR equation
##### h(x)= theta_0 + theta_1 * x1 + theta_2 * x2 + theta_3 * x3

```python
learning_rate = 0.00001
maxNumOfIteration = 1000
thetas0, thetas1, thetas2, thetas3, loss, hyposesis, optimum, numIter = GDBatchMulVar(x_points, y_points, learning_rate, maxNumOfIteration)

print('thetas0: ', thetas0)
print('thetas1: ', thetas1)
print('thetas2: ', thetas2)
print('thetas3: ', thetas3)
print('loss: ', loss)
print('hyposesis: ', hyposesis)
print('optimum: ', optimum)
```

#### Use R2 score to evaluate LR equation output

```python
from sklearn.metrics import r2_score
r2_score(y_points, optimum)
```
